Suppose that fraction $t_e$ of the land is enclosed.  

Laborers are free to move between enclosed and customary sectors in search of the highest return.  In the unenclosed or customary sector a laborer will earn the average product of labor $APL_c$ while in the enclosed sector they earn the prevailng market wage $w_e$ given by the value marginal product of labor in that sector. Hence, in equilibrium:

$$
APL_c = w_e
$$

With competitive factor markets, the market wage $w_e$ is given by the value marginal product of labor on enclosed farms so $w_e = MPL_e$.  By Euler's Theorem 

$$
F(T_c, L_c) = MPL_c \cdot L_c + MPT_c \cdot T_c
$$

Hence we can write the average product of labor in the customary sector as:

$$
APL_c = \frac{F(T_c, L_c)}{L_c} = MPL_c + MPT_c \cdot \frac{T_c}{L_c}
$$

We can think of the customary sector as having competitive factor markets. In equilibrium laborers are indifferent between moving to the enclosed sector to earn a market wage or remaining in the customary sector where they can earn a lower wage for their labor and earn rents from the land they own by possession.  We can even assume that, from this equilibrium, land and labor are traded on locally competitive factor markets. The prevailing local wage is $w_c = MPL_c$ and the prevailing rental rate of customary land $r_c = MPT_c$.  In equilibrium we can therefore write:

$$
APL_c= w_c + r_c \cdot \frac{T_c}{L_c}
$$

which implies that in equilibrium:

$$
w_c + r_c \cdot \frac{T_c}{L_c}  = w_e
$$

Re-arranging, we are led to the conclusion that, in equilibrium:

$$
r_c \cdot \frac{T_c}{L_c}  = w_e - w_c  
$$

This makes perfect sense: people enter (remain in) the customary sector so long as the value of the land rents they can capture via possesion and continued membership in the community is exactly offset by the opportunity cost of doing so (the wages they must forgo by remaining to establish and protect their land claims).


We can see this in the following diagram:

```python
plotmpts(te = 1/2, alp= 1/2, th = 1.3, tlbar = 1);
```
